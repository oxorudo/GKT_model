### data_all , train, test, valid 아이스크림에듀 처럼 데이터 형태 변환

```python
import pandas as pd

def analyze_student_performance_and_save(df):
    # 전체 학습자 선택
    all_users = df['UserID'].drop_duplicates()

    # 최종 결과를 저장할 리스트
    result_list = []

    # 전체 학습자별로 반복
    for user_id in all_users:
        # 해당 학습자가 푼 문제 필터링
        student_data = df[df['UserID'] == user_id]

        # 'O'와 'X' 값만 있는지 확인하고 필터링
        filtered_data = student_data[student_data['Correct'].isin(['O', 'X'])]

        # 맞으면 1, 틀리면 0으로 변환
        filtered_data['정오'] = filtered_data['Correct'].map(lambda x: 1 if x == 'O' else 0)

        # knowledgeTag와 정오 값을 리스트로 추가
        knowledge_tags = filtered_data['knowledgeTag'].tolist()
        correct_values = filtered_data['정오'].tolist()

        # 먼저 학습자의 ID를 따로 추가
        result_list.append([user_id])

        # 그 다음 knowledgeTag와 정오 결과를 추가
        result_list.append(knowledge_tags)
        result_list.append(correct_values)

    # DataFrame으로 변환, 컬럼 이름 없이 저장
    result_df = pd.DataFrame(result_list)

    # 결과를 train.csv로 저장 (header=False로 칼럼명 없앰)
    result_df.to_csv('train.csv', index=False, header=False, encoding='utf-8-sig')

    return result_df


# 주어진 데이터를 이용하여 train.csv 생성
result_df = analyze_student_performance_and_save(student_df_12)

# 결과 출력
result_df
```

#### train, test, valid

```python
# Select rows with index multiples of 3 (0, 3, 6, ...)
filtered_rows = data_all[data_all.index % 3 == 0]

# Extract the first column values
first_column_values = filtered_rows.iloc[:, 0].values

# Step 1: Split the entire dataset into train and test sets (80:20)
train_size = int(len(first_column_values) * 0.8)
train_ids = first_column_values[:train_size]  # 80% for train
test_ids = first_column_values[train_size:]   # 20% for test

# Step 2: Further split the train set into train and valid sets (80:20)
valid_size = int(len(train_ids) * 0.2)
train_ids_final = train_ids[:-valid_size]  # 80% of train set for training
valid_ids = train_ids[-valid_size:]         # 20% of train set for validation

# Step 3: Create extended lists by capturing the rows from data_all
train_ids_extended = []
valid_ids_extended = []
test_ids_extended = []

# For train set (80% of first_column_values)
for value in train_ids_final:
    index = filtered_rows[filtered_rows.iloc[:, 0] == value].index[0]  # Get the original index
    train_ids_extended.append(data_all.loc[index:index + 2])  # Get the current row and the next two

# For valid set (20% of the train set)
for value in valid_ids:
    index = filtered_rows[filtered_rows.iloc[:, 0] == value].index[0]  # Get the original index
    valid_ids_extended.append(data_all.loc[index:index + 2])  # Get the current row and the next two

# For test set (20% of first_column_values)
for value in test_ids:
    index = filtered_rows[filtered_rows.iloc[:, 0] == value].index[0]  # Get the original index
    test_ids_extended.append(data_all.loc[index:index + 2])  # Get the current row and the next two

# Step 4: Combine the extended lists into DataFrames
train_df = pd.concat(train_ids_extended)
valid_df = pd.concat(valid_ids_extended)
test_df = pd.concat(test_ids_extended)
```

```python
# Step 5: Save to CSV
train_df.to_csv('train_split_with_rows.csv', index=False, encoding='utf-8-sig', header=False)
valid_df.to_csv('valid_split_with_rows.csv', index=False, encoding='utf-8-sig', header=False)
test_df.to_csv('test_split_with_rows.csv', index=False, encoding='utf-8-sig', header=False)

print("CSV files saved with associated rows: train_split_with_rows.csv, valid_split_with_rows.csv, test_split_with_rows.csv")
```
