```python
import pandas as pd

df = pd.read_parquet("data/filtered_combined_user_data.parquet")
df2 = pd.read_parquet("data/filtered_combined_user_data1.parquet")

# 비교할 컬럼 목록
compare_columns = ["UserID", "QuizCode", "CreDate", "Correct"]

# 두 데이터프레임에서 동일한 행 필터링
df1_filtered = df[
    df[compare_columns]
    .apply(tuple, axis=1)
    .isin(df2[compare_columns].apply(tuple, axis=1))
]
df2_filtered = df2[
    df2[compare_columns]
    .apply(tuple, axis=1)
    .isin(df[compare_columns].apply(tuple, axis=1))
]
```
